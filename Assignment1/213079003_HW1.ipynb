#### (a) Generate sinusoidal sequences with $ω_{0}$ = 0.3$\pi$ and $ω_{0}$ = 0.75$\pi$. Determine the period of each sequence from the plot and verify the result theoretically.


```python
# 10a Plot of Sinusoidal Sequences
def sinusoid(A,w0,n,phi):
    sinusoid = []
    for i in n:
        sinusoid.append(A* np.sin((w0*i) + phi))
    return sinusoid
n = np.arange(-10,25,1)

x1 = sinusoid(1,0.3*np.pi,n,0)
x2 = sinusoid(1,0.75*np.pi,n,0)
figure = plt.figure(figsize=(15, 12))
plt.subplot(2,1,1)
plt.stem(n,x1)
plt.xlabel('n',fontsize="14")
plt.ylabel('$x_{1}[n]$',fontsize="14")
plt.title("$x_{1}[n]$ = sin(0.3πn)",fontsize="16")
plt.grid(True)
plt.subplot(2,1,2)
plt.stem(n,x2)
plt.xlabel('n',fontsize="14")
plt.ylabel('$x_{2}[n]$',fontsize="14")
plt.title("$x_{2}[n]$ =  sin(0.75πn)",fontsize="16")
plt.grid(True)
plt.suptitle('Plot of Sinusoidal Sequences', color='red',fontsize="18")
figure.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

##### OBSERVATION
For a sequence x[n]= sin($w_{0}$n), the period is given by

N = $\frac{2πm}{w_{0}}$

where m, N are integers.

For the sequence $x_{1}[n]$ = sin(0.3πn), the period is given by

N = $\frac{2πm}{w_{0}}$ =$\frac{2πm}{0.3π}$ = $\frac{20m}{3}$

For N to be an integer, m=3. Thus the fundamental period of $x_{1}[n]$ = sin(0.3πn) is N=20sec. 

For the sequence $x_{2}[n]$ =  sin(0.75πn), the period is given by

N = $\frac{2πm}{w_{0}}$ =$\frac{2πm}{0.75π}$ = $\frac{8m}{3}$

For N to be an integer, m=3. Thus the fundamental period of $x_{2}[n]$ =  sin(0.75πn) is N=8sec. 

This is verified in the plots also.

From 11a,11b it is seen that as the number of samples N increases, the magnitude spectrum is decreasing at a rapid rate.

Also in 11c, since the spectrum is a square value, giving large N makes the spectrum to decay at a rapid rate.

```python
def tri(N):
  n=np.arange(-N,N+1,1)
  x=[]
  for i in n:
    if i<0:
      value= 1 + (i/N)
      x.append(value)
    else:
      value= 1 - (i/N)
      x.append(value)
  plt.stem(n,x)
  return x
y=tri(5)

```
